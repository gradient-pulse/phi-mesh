```python
import numpy as np, matplotlib.pyplot as plt

t = np.linspace(0, 20, 400)
g = np.sin(t) + 0.4*np.sin(3*t + 0.8) + 0.25*np.random.randn(len(t))

dt = t[1] - t[0]
dZ = np.gradient(g, dt)
sigma = dZ.std()

nt = np.where(
    (np.sign(dZ[:-1]) != np.sign(dZ[1:])) &
    (np.abs(dZ[1:]) > 1.5 * sigma)
)[0]

plt.plot(t, g, lw=1.2)
plt.scatter(t[nt], g[nt], c="red", label="Narrative Ticks")
plt.xlabel("Time"); plt.ylabel("G(t)")
plt.title("Synthetic NT detection"); plt.legend(); plt.show()
```
